# Interpretation (4)

Describe something interesting about your corpus that you discovered during the process of
completing this assignment.
At a minumum, use 250 words, but you may use more. You may also add images if you’d like.

One thing that both interested and surprised me about my corpus was the large variety of language among artists within the same genre of music. Based on my preexisting knowledge of music, these artists mostly fall into the pop category, with a few being more pop-adjacent. I figured that artists whose music falls into the same genre would generally have very similar themes and language. I thought many of the terms would be related to love and relationships, as that is a frequent topic in pop music. However, I discovered that there is a lot more diversity among these artists’ lyrics than I originally thought. Specifically, I found that artists who are more similar to each other on the surface exhibit subtle but meaningful differences. An example of this is illustrated through analyzing Beyonce and Rihanna. Their music falls within the pop/R&B genre, and they have a lot of overlapping themes as well as similar listeners. My analysis showed that both artists have similar raw sentiment values (slightly positive), but I was surprised to see that their topic distributions differed. Beyonce has higher values in T01 (dance/attraction) whereas Rihanna has higher values in T03 (love). I would have assumed that both artists would have more overlap in topic emphasis given the similarity among their music. This comparison highlights the fact that artists whose music has similar thematic components might differ in their emphasis of those themes in the structure of their lyrics. Overall, my preexisting knowledge of these artists' music was different from what this analysis showed, illustrating the high impact that text analytics can have in discovering subtle differences.
